# FIND OCCURRENCE OF THE EMPLOYEE NAME(TOP5)

```python
uq = data['JobTitle'].nunique()
print(uq)
```

# TOTAL NUMBER OF JOBS TITLES CONTAINS CAPTAIN

```python
captain_titles_count = data[data['JobTitle'].str.contains('Captain', case=False, na=False)].shape[0]
print("Total number of job titles containing 'Captain':", captain_titles_count)
```

# DISPLAY ALL THE EMPLOYEE NAMES FROM FIRE DEPARTMENT

```python
fire_department_employees = data[data['JobTitle'].str.contains('FIRE DEPARTMENT', case=False, na=False)]
fire_department_employee_names = fire_department_employees['EmployeeName'].unique()
print("Employee names from FIRE DEPARTMENT:")
print(fire_department_employee_names)
```

# FIND MINIMUM,MAXIMUM AND AVERGE BASEPAY

```python
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')
min_basepay = data['BasePay'].min()
max_basepay = data['BasePay'].max()
avg_basepay = data['BasePay'].mean()
print(min_basepay)
print(max_basepay)
print(avg_basepay)
```

# REPLACE 'NOT PROVIDED' IN  EMPLOYEENAME COLUMN TO NaN

```python
data['EmployeeName'].replace('NOT PROVIDED', pd.NA, inplace=True)
```

# drop the rows having more than 5 missing values

```python
threshold = 5
data_cleaned = data.dropna(thresh=len(data.columns) - threshold)                        
```

# find the job title of ALBERT PARDINI

```python
albert_pardini_job_title = data.loc[data['EmployeeName'] == 'ALBERT PARDINI', 'JobTitle'].iloc[0]
print("Job Title of Albert Pardini:", albert_pardini_job_title)
```

# HOW MUCH ALBERT PARDINI MAKE (INCULDE DENEFITS)?

```python
albert_pardini_total_pay_benefits = data.loc[data['EmployeeName'] == 'ALBERT PARDINI', 'TotalPayBenefits'].iloc[0]
print("Total Pay and Benefits of Albert Pardini:", albert_pardini_total_pay_benefits)
```

# DISPLAY NAME OF THE PERSON HAVINNG THE HIGHEST BASEPAY

```python
highest_basepay_employee = data.loc[data['BasePay'].idxmax()]['EmployeeName']
print("Name of the person with the highest BasePay:", highest_basepay_employee)
```

# FIND AVERGE BASEPAY OF ALL EMPLOYEE PER YEAR

```python
average_basepay_per_year = data.groupby('Year')['BasePay'].mean()
print("Average BasePay per Year:")
print(average_basepay_per_year)
```

# FIND AVERGE BASEPAY OF ALL EMPLOYEE PER JOB TITLE

```python
average_basepay_per_job_title = data.groupby('JobTitle')['BasePay'].mean()
print("Average BasePay per Job Title:")
print(average_basepay_per_job_title)
```

# FIND AVERGE BASEPAY OF EMPLOYEE HAVIG JOB TITLE ACCOUNTANT

```python
average_basepay_accountant = data[data['JobTitle'] == 'ACCOUNTANT']['BasePay'].mean()
print("Average BasePay for employees with the job title 'ACCOUNTANT':", average_basepay_accountant)
```

# FIND TOP5 MOST COMMAN JOBS

```python
top5_common_jobs = data['JobTitle'].value_counts().head(5)
print("Top 5 most common job titles:")
print(top5_common_jobs)
```
